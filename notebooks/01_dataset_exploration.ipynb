```python
import sys
from pathlib import Path

import numpy as np
import pandas as pd

sys.path.append(str(Path().resolve().parent / "src"))
from utils import quick_overview
```

```python
def load_all_csvs(folder_path: str = "data"):
    folder = Path.cwd().parent / folder_path
    csv_files = list(folder.glob("*.csv"))
    dfs = {f.stem: pd.read_csv(f) for f in csv_files}
    return dfs
```

```python
dfs = load_all_csvs()
dfs.keys()
```
